```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset

df = pd.read_csv("../data/raw/data.csv")
df.head()
```

## Data Cleaning

```python
#Convert all column names to lowercase for consistency
def lowercase_column_names(df):
    """
    Conver all column names of a DataFrame to lowercase.

    Parameters:
    df: The DataFrame whose column names need to be converted

    Returns:
    df: The DataFrame with all column names in lowercase
    """
    df = df.rename(columns=str.lower)
    return df
```

```python
df = lowercase_column_names(df)
df.head()
```

```python
#rename 'msrp' column to 'price
df = df.rename(columns={'msrp':'price'})
df.head()
```

```python
#display any missing values in the dataset
df.isnull().sum()
```

```python
#Defining a function to clean the columns containing missing values
def missingvalues_df(df):
    
    #drop row with missing values on Market Catefory
    df.dropna(subset='market category',inplace = True)
    
    # columns with mean
    df['engine hp'].fillna(df['engine hp'].mean(), inplace=True)
    df['engine cylinders'].fillna(df['engine cylinders'].mean(), inplace=True)

    # column with mode
    df['number of doors'].fillna(df['number of doors'].mode().iloc[0], inplace=True)
    return df
```

```python
df_cars = missingvalues_df(df)
```

```python
#checking again whether there are any remaining missing values
df_cars.isnull().sum()
```

```python
df_cars.head()
```

## Exploratory Data Analysis

```python
#Generate a color palette with the same number of uniques values for 'make'
num_colors = df_cars['make'].nunique()
palette = sns.color_palette("husl", num_colors)

#get the value counts of the 'make' columns and sort by count
order = df_cars['make'].value_counts().index

#Create a countplot for each car brand in the dataset
plt.figure(figsize = (20,15))
sns.countplot(y=df_cars['make'], order = order, palette = palette)
plt.title("Number of Cars Manufactured by Brand")
plt.show()
```

```python
#Generate a color palette with the same number of uniques values for 'year'
num_colors_year = df_cars['year'].nunique()
palette_year = sns.color_palette("husl", num_colors_year)

#Create a countplot for each year in the dataset
plt.figure(figsize = (20,15))
sns.countplot(x=df_cars['year'], palette = palette_year)
plt.title("Number of Cars by Year")
plt.show()
```

```python
#Generate a color palette with the same number of uniques values for 'engine fuel type'
num_colors_type = df_cars['engine fuel type'].nunique()
palette_type = sns.color_palette("husl", num_colors_type)

#get the value counts of the 'engine fuel type' columns and sort by count
order_type = df_cars['engine fuel type'].value_counts().index

#Create a countplot for each fuel type in the dataset
plt.figure(figsize = (18,10))
sns.countplot(y=df_cars['engine fuel type'], order = order_type, palette = palette_type)
plt.show()
```

```python
#Generate a color palette with the same number of uniques values for 'transmission type'
num_colors_transmission = df_cars['transmission type'].nunique()
palette_transmission = sns.color_palette("husl", num_colors_transmission)

#get the value counts of the 'tranmission type' columns and sort by count
order_transmission = df_cars['transmission type'].value_counts().index

#Create a countplot for each type in the dataset
plt.figure(figsize = (18,10))
ax = sns.countplot(x=df_cars['transmission type'], order = order_transmission, palette = palette_transmission)
new_labels = ['Automatic', 'Manual', 'Automated Manual', 'Direct Drive', 'Unknown']
ax.set_xticklabels(new_labels)
plt.show()
```

```python
#Generate a color palette with the same number of uniques values for 'vehicle size'
num_colors_size = df_cars['vehicle size'].nunique()
palette_size = sns.color_palette("husl", num_colors_size)

#get the value counts of the 'vehicle size' columns and sort by count
order_size = df_cars['vehicle size'].value_counts().index

#Create a countplot for each size in the dataset
plt.figure(figsize = (18,10))
sns.countplot(x=df_cars['vehicle size'], order = order_size, palette = palette_size)
plt.show()
```

```python
#Generate a color palette with the same number of uniques values for 'vehicle style'
num_colors_style = df_cars['vehicle style'].nunique()
palette_style = sns.color_palette("husl", num_colors_style)

#get the value counts of the 'vehicle style' columns and sort by count
order_style = df_cars['vehicle style'].value_counts().index

#Create a countplot for each style in the dataset
plt.figure(figsize = (18,10))
sns.countplot(y=df_cars['vehicle style'], order = order_style, palette = palette_style)
plt.show()
```

```python
#Calculate mean price per make
make_mean = df_cars.groupby('make')['price'].mean().reset_index()
make_mean_sorted = make_mean.sort_values(by='price', ascending = False)

#Plot the mean price per make
plt.figure(figsize = (20,12))
sns.barplot(data=make_mean, x ='make', y='price', order=make_mean_sorted['make'])
plt.title("Mean price of Cars per Brand")
plt.xlabel('Brand')
plt.xticks(rotation = 90)
plt.ylabel('Mean Price')
plt.show()
```

```python
#Calculate mean price per transmission type
type_mean = df_cars.groupby('transmission type')['price'].mean().reset_index()

#Plot the mean price per transmission type
plt.figure(figsize = (20,12))
ax = sns.barplot(data=type_mean, x ='transmission type', y='price')
plt.title("Mean price of Cars per Transmission Type")
plt.xlabel('Transmission Type')
plt.ylabel('Mean Price')
new_label = ['Automated Manual', 'Automatic', 'Direct Drive', 'Manual', 'Unknown']
ax.set_xticklabels(new_label)
plt.show()
```

```python
#Calculate mean price per driven_wheels
wheels_mean = df_cars.groupby('driven_wheels')['price'].mean().reset_index()
wheels_mean_sorted = wheels_mean.sort_values(by='price', ascending = False)
#Plot the mean price per driven_wheels
plt.figure(figsize = (20,12))
sns.barplot(data=wheels_mean, x ='driven_wheels', y='price', order = wheels_mean_sorted['driven_wheels'])
plt.title("Mean price of Cars per Driven Wheels")
plt.xlabel('Driven Wheels')
plt.ylabel('Mean Price')
plt.show()
```

```python
#Calculate popularity of each car brand
make_pop = df_cars.groupby('make')['popularity'].mean().reset_index()
make_pop_sorted = make_pop.sort_values(by='popularity', ascending = False)

#Plot the mean price per make
plt.figure(figsize = (20,15))
sns.barplot(data=make_pop, y ='make', x='popularity', order=make_pop_sorted['make'], palette = palette)
plt.title("Popularity of each Car Brand")
plt.ylabel('Brand')
plt.xlabel('Mean Price')
plt.show()
```
